```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

```python
df = pd.read_csv(r'C:\Users\yesia\Downloads\BostonHousing.csv')
```

```python
#  making heatman to understand correlation

plt.figure(figsize=(10,10))
sns.heatmap(cor, cbar=True, square=True,annot=True,fmt='.1f',annot_kws={'size':8}, cmap='Blues')
```

```python
#  splitting the data 

X= df.drop(['medv'], axis=1)
Y= df['medv']
```

```python
print(X)
print(Y)
```

```python
# splitting the data into training data and test data

X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size=0.2, random_state=2)
```

```python
print(X_train,X_test)
```

```python
print(X.shape,X_train.shape,X_test.shape,Y_train.shape,Y_test.shape)
```

```python
# model training

# XGBoost Regressor

# loading the model

model = XGBRegressor()
```

```python
# training the model with x_train

model.fit(X_train,Y_train)
```

```python
# evaluation
# prediction on training data
# accuracy on prediction on training data

training_data_prediction = model.predict(X_train)
```

```python
print(training_data_prediction)
```

```python
# R square error method

score_1 = metrics.r2_score(Y_train, training_data_prediction)

# mean absolute error

score_2 = metrics.mean_absolute_error(Y_train, training_data_prediction)

print("R Score Error:",score_1)
print("Mean Absolute Error:", score_2)
```

```python
#  visualizing the actual price and predicted prices

plt.scatter(Y_train,training_data_prediction)
plt.xlabel('ActualPrice')
plt.ylabel('PredictedPrice')
plt.title('ActualvsPredicted')
plt.show()
```

```python
#  we need to find the value for test data also

test_data_prediction = model.predict(X_test)

# R square error method

score_3 = metrics.r2_score(Y_test, test_data_prediction)

# mean absolute error

score_4 = metrics.mean_absolute_error(Y_test, test_data_prediction)

print("R Score Error:",score_3)
print("Mean Absolute Error:", score_4)
```

```python
#  visualizing the actual price and predicted prices

plt.scatter(Y_test,test_data_prediction)
plt.xlabel('ActualPrice')
plt.ylabel('PredictedPrice')
plt.title('ActualvsPredicted')
plt.show()
```
