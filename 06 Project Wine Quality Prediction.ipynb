```python
# data preprocessing
# seperating data and label
X = df.drop('quality', axis=1)
```

```python
print(X)
```

```python
#  we will binarise the value 


Y= df['quality'].apply(lambda y_value: 1 if y_value>=7 else 0)
```

```python
print(Y)
```

```python
# splitting the data into train and test

X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=3)
```

```python
print(Y.shape, Y_train.shape, Y_test.shape)
```

```python
# model training for which we will use random forest classifier

model = RandomForestClassifier()
```

```python
 model.fit(X_train,Y_train)
```

```python
#  model_evaluation

# accuracy_score

# accuracy on test data

X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print("Accuracy Score :", test_data_accuracy)
```

```python
# building a predictive system

input_data = (7.8,0.61,0.29,1.6,0.114,9.0,29.0,0.9974,3.26,1.56,9.1)

# changing the input data into numpy array

input_data_as_numpy_array = np.asarray(input_data)

# reshape the data as we are predicting the label for one instance

input_data_reshaped= input_data_as_numpy_array.reshape(1,-1)

prediction = model.predict(input_data_reshaped)
print(prediction)

if (prediction[0]==1):
    print("Good Quality Wine")
else:
    print("Bad Quality Wine")
```
