```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# checking the distribution of legit transaction and fruad transaction

df['Class'].value_counts()
```

```python
#  0 = normal transaction
#  1  = fruad transaction
#  this data is highly unbalanced
```

```python
#  seperating data for analysis

legit = df[df.Class == 0 ]
fruad = df[df.Class == 1 ]
```

```python
print(legit.shape)
print(fruad.shape)
```

```python
#  now we will build sample data set containing similar distribution of normal and fruad transaction
#  here we will take 492 random transaction so that we can have both fraud and legit transaction equal data
```

```python
legit_sample = legit.sample(n=492)
```

```python
# concatinating two dataframe

newdata=pd.concat([legit_sample, fruad], axis=0)
```

```python
 newdata.head()
```

```python
newdata['Class'].value_counts()
```

```python
newdata.groupby('Class').mean()
```

```python
# splitting the data

X = newdata.drop(columns= 'Class' , axis = 1)
Y= newdata['Class']
```

```python
# train_test_split

X_train, X_test, Y_train,Y_test = train_test_split(X,Y, test_size=0.2, random_state=2, stratify = Y)
```

```python
print(X.shape, X_train.shape,X_test.shape)
```

```python
# model training

model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
# model evaluation and finding the value

X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Model Accuracy:', training_data_accuracy)
```

```python
#  for test data 

X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print('Model Accuracy:', test_data_accuracy)
```
